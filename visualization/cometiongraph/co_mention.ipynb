```python
begin = time.time()
for text in tqdm(df['text']):
    words = set()
    for word in text.split(' '):
        if word in wordDict:
            words.add(word)
    words = list(words)
    for i in range(len(words)):
        for j in range(i+1, len(words)):
            graph[i][j] += 1
            graph[j][i] += 1
print('Time Elapsed', str(round(time.time() - begin, 2)) + 's')
```

```python
begin = time.time()
nodes1 = []
nodes2 = []
values = []
for i in tqdm(range(k)):
    for j in range(i+1, k):
        if graph[i][j] > 0:
            nodes1.append(wordList[i])
            nodes2.append(wordList[j])
            values.append(graph[i][j])
print('Time Elapsed', str(round(time.time() - begin, 2)) + 's')            
```
